# Яндекс. Ночь анализа данных.

```python
%matplotlib inline

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## Задача для учебного потока

Загрузим данные

```python
path_to_data = 'data.tsv'
data_df = pd.read_csv(path_to_data, sep='\t', names=['TIME', 'ID', 'TYPE', 'ROUTE', 'X', 'Y'])

data_df.head()
```

Изобразим все координаты

```python
plt.figure(figsize = (10, 10))
plt.scatter(data_df.X, data_df.Y)
plt.legend(['gps'])
```

Изобразим координаты одного маршрута, например с номером 137

```python
route = data_df[data_df['ROUTE'] == 137]

plt.figure(figsize = (10, 10))
plt.scatter(route.X, route.Y)
plt.legend(['gps'])
```
